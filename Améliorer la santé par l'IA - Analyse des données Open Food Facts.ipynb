```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
#A function to unfold a string as a liste
def RemoveTag(strInput):
    position = strInput.find(":")
    if(position!=-1):
        strInput = strInput[0:position]
def UnfoldList(x,removeTag = False,removeDoublon=True,separator=","):
    liste = str(x).split(separator)
    if removeTag:
        for i in range(len(liste)):
            position = liste[i].find(":")
            if(position!=-1):
                liste[i] = liste[i][0:position]
    if removeDoublon:
        liste = list(set(liste))
    return liste


def SetToCount(x:list)->int:
    if(type(x)!=list):
        return None
    return len(x)

def ToCount(x,remTag = False,remDoublon=True,separator=","):
    list = UnfoldList(x,removeTag=remTag,removeDoublon=remDoublon)
    return SetToCount(list)

def CreateHeatMapCorr(collumns,baseDf,title="Matrice de corrélation"):
    dfChoice = baseDf[collumns]
    matrice = dfChoice.corr()
    np.fill_diagonal(matrice.values,np.nan)
    #mask = np.triu(np.ones_like(matrice, dtype=bool))
    #plt.title("Matrice de corrélation")
# Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(matrice,mask=np.isnan(matrice), cmap=cmap, center=0,
            square=True,cbar_kws={"shrink": .5},linewidths=.5,annot=True,linecolor="black").set_title(title)
#

def GetCorrelationPair(dataFrame,correlationRequired):
    matrice = dataFrame.corr()
    triu_indices = np.triu_indices_from(matrice, k=0)
    # Set elements above the diagonal to zero
    matrice.values[triu_indices] = 0
    #print(matrice)
    for col in matrice:
        row = 0
        for cell in matrice[col]:
            #print(f"Cell :{cell},Row {row}")
            if(cell>correlationRequired):
                print(f"{col} est fortement corrélé avec {matrice.index[row]}")
                #print(row)
            row+=1
    
```

```python
#11. Comment se calcule le Nutri-Score (FR et UK) ? Qu'est-ce qui les différencie ?
#Le nutriscore uk n'existe pas l'agence indiquer dans les meta données : le FSA
#CreateHeatMapCorr(["nutrition-score-fr_100g","nutrition-score-uk_100g"],df)

dfAdit = dfClean.filter(like='_100g',axis=1)
dfAdit["nutrition-score-fr_100g"]=dfClean["nutrition-score-fr_100g"]
dfAdit["nutrition-score-uk_100g"]=dfClean["nutrition-score-uk_100g"]

correlation = dfAdit.corr()
nutriCorr = correlation[["nutrition-score-uk_100g","nutrition-score-fr_100g"]]


#aditCorr_sorted = dataframeadit.reindex(aditCorr.abs().sort_values(ascending=False).index)
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(60, 20))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(nutriCorr.T, cmap=cmap, center=0,
            square=True, linewidths=.5,xticklabels=nutriCorr.index,yticklabels=["nutri-UK","nutri-FR"], cbar_kws={"shrink": .5},annot=True)
dfAdit
```
