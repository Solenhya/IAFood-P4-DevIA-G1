```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
#A function to unfold a string as a liste
def RemoveTag(strInput):
    position = strInput.find(":")
    if(position!=-1):
        strInput = strInput[0:position]
def UnfoldList(x,removeTag = False,removeDoublon=True,separator=","):
    liste = str(x).split(separator)
    if removeTag:
        for i in range(len(liste)):
            position = liste[i].find(":")
            if(position!=-1):
                liste[i] = liste[i][0:position]
    if removeDoublon:
        liste = list(set(liste))
    return liste


def SetToCount(x:list)->int:
    if(type(x)!=list):
        return None
    return len(x)

def ToCount(x,remTag = False,remDoublon=True,separator=","):
    list = UnfoldList(x,removeTag=remTag,removeDoublon=remDoublon)
    return SetToCount(list)
```

```python
#Question 10 
# Existe-t-il une corrélation entre la présence d'additifs et la valeur nutritionnelle des aliments ?
dfAdit = df.filter(like='_100g',axis=1)
dfAdit["additive_n"]=df["additives_n"]
correlation = dfAdit.corr()
aditCorr = correlation["additive_n"]
dataframeadit = pd.DataFrame(aditCorr)
aditCorr_sorted = dataframeadit.reindex(aditCorr.abs().sort_values(ascending=False).index)
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(60, 20))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(aditCorr_sorted.T, cmap=cmap, center=0,
            square=True, linewidths=.5,xticklabels=aditCorr_sorted.index,yticklabels=["Additives"], cbar_kws={"shrink": .5},annot=True)
```

```python
#11. Comment se calcule le Nutri-Score (FR et UK) ? Qu'est-ce qui les différencie ?
#Le nutriscore uk n'existe pas l'agence indiquer dans les meta données : le FSA
#CreateHeatMapCorr(["nutrition-score-fr_100g","nutrition-score-uk_100g"],df)

dfAdit = df.filter(like='_100g',axis=1)
dfAdit["nutrition-score-fr_100g"]=df["nutrition-score-fr_100g"]
dfAdit["nutrition-score-uk_100g"]=df["nutrition-score-uk_100g"]

correlation = dfAdit.corr()
nutriCorr = correlation[["nutrition-score-uk_100g","nutrition-score-fr_100g"]]


#aditCorr_sorted = dataframeadit.reindex(aditCorr.abs().sort_values(ascending=False).index)
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(60, 20))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(nutriCorr.T, cmap=cmap, center=0,
            square=True, linewidths=.5,xticklabels=nutriCorr.index,yticklabels=["nutri-UK","nutri-FR"], cbar_kws={"shrink": .5},annot=True)
dfAdit
```

```python
nutriCorrDif = nutriCorr
nutriCorrDif["difference"] = nutriCorrDif["nutrition-score-uk_100g"]-nutriCorrDif["nutrition-score-fr_100g"]
nutriCorrDif = nutriCorrDif.sort_values(by="difference",ascending=False)
f, ax = plt.subplots(figsize=(60, 20))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(nutriCorrDif.T, cmap=cmap, center=0,
            square=True, linewidths=.5,xticklabels=nutriCorrDif.index,yticklabels=["nutri-UK","nutri-FR","Difference"], cbar_kws={"shrink": .5},annot=True)
```
